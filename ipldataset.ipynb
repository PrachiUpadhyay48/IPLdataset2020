```python
# seaborn bar plot:

import pandas as pd
sns.set(style="whitegrid")
data=pd.read_csv("data2.csv")
```

```python
sns.barplot(x="length",y="signal",hue="signal",data=data)
plt.show()
```

```python
# seaborn scatterplot:
sns.scatterplot(x="timepoint",y="length",hue="timepoint",s=300,data=data)
plt.show()
```

```python
# seaborn histogram/displot:

data=pd.read_csv("data2.csv")
data.head()
```

```python
# histogram:
sns.displot(data['timepoint'])
plt.show()
```

```python
#kdeplot:(univarient or bivariant distribution)

sns.displot(data['timepoint'],kind="kde")
plt.show()
```

```python
#ecdfplot:

sns.displot(data['length'],kind="ecdf")
plt.show()
```

```python
# kde curve:

sns.displot(data['length'],kde=True)
plt.show()
```

```python
# to draw bivariant plot:

sns.displot(data=df,x='width',y="timepoint")
plt.show()
```

```python
# kde

sns.displot(data=df,x='width',y="timepoint",kind="kde")
plt.show()
```

```python
# kde with each rug

r=sns.displot(data=df,x='length',y="timepoint",kind="kde",rug=True)
plt.show()
```

```python
# hue mapping:

sns.displot(data=df,x='width',hue="region",kind="kde")
plt.show()
```

```python
# plotting: (further customerization)

sns.displot(data=df,x='size',hue="region",color='g',multiple="stack")
plt.show()
```

```python
#facetgrid (subsets of distinct subplots or"facets"):

sns.displot(data=df,x='width',hue="region",col='sex',kind="kde",height=3)
```

```python
# facetgrid with its size nd height aspect:

sns.displot(data=df,y="timepoint",hue='region',col="sex",kind="ecdf",height=2,aspect=5)
plt.show()
```

```python
#facetgrid with further customerize:

r=sns.displot(data=df,y="length",hue="sex",col="region",kind="kde",height=2,aspect=2)
plt.show()
```

```python
#seaborn jointplot:

data=pd.read_csv("data2.csv")
data.head()
```

```python
# jointplot with regression:

sns.jointplot(x="length",y="size",data=data,color="olive",kind="reg")
plt.show()
```

```python
#box-plot:

sns.boxplot(x="event",y="timepoint",data=data)
plt.show()
```

```python
#box-plot palette:

sns.boxplot(x="region",y="length",data=data,hue="sex",palette="Spectral_r")
plt.show()
```
